# Downloading and Reading Files

```r
# download csv
# https://data.baltimorecity.gov/api/views/aqgr-xx9h/rows.csv?accessType=DOWNLOAD
if(file.exists("05_01_downloading_data_cameras.csv")) {
    "file already downloaded"
} else {
    file_url <- "https://data.baltimorecity.gov/api/views/aqgr-xx9h/rows.csv?accessType=DOWNLOAD"
    download.file(file_url, destfile="05_01_downloading_data_cameras.csv")
    if(file.exists("05_01_downloading_data_cameras.csv")) {
        "file download correctly"
    } else {
        "file not downloaded correctly"
    }
}
```

```r
# read.csv has ,sep = ",", header = TRUE by default
d2 <-read.csv("05_01_downloading_data_cameras.csv")
head(d2)
```

```r
source <- "https://www.w3schools.com/xml/simple.xml"
dest <- "05_01_downloading_data_climate_simple.xml"

if(file.exists(dest)) {
    "file already downloaded"
} else {
    file_url <- source
    download.file(file_url, destfile=dest)
    if(file.exists(dest)) {
        "file download correctly"
    } else {
        "file not downloaded correctly"
    }
}
```

```r
# in ubuntu you may need to run the following commands

# in linux shell
# > sudo apt-get update
# > sudo apt-get install libxml2-dev

# in REPL
# > install.packages("XML", dependencies = TRUE)
library(XML)
```

```r
d4 <- xmlTreeParse("05_01_downloading_data_climate_simple.xml", useInternalNodes = TRUE)
rootNode <- xmlRoot(d4)
xmlName(rootNode)
names(rootNode)
rootNode[[1]]
rootNode[[1]][[1]]
```

```r
# extracting values from a XML
d5 <- xmlTreeParse("05_01_downloading_data_climate_simple.xml", useInternalNodes = TRUE)
rootNode <- xmlRoot(d5)
xmlValue <- function(node) {
    name <- node[[1]][[1]] # name
    price <- node[[2]][[1]] # price
    desc <- node[[3]][[1]] # description
    calories <- node[[4]][[1]] # calories
    c(name, price, desc, calories)
}
xpathApply(rootNode,"food", xmlValue)
```
